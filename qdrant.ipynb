```python
from qdrant_client import QdrantClient
from qdrant_client.models import Distance, VectorParams

client = QdrantClient("localhost", port=6333)

client.create_collection(
    collection_name="documents",
    vectors_config=VectorParams(
        size=4,            # vector dimension
        distance=Distance.COSINE
    )
)

print("Collection created!")
```

```python
from qdrant_client.models import PointStruct

points = [
    PointStruct(
        id=7,
        vector=[0.0, 0.2, 0.3, 0.4],
        payload={
            "text": "Python is a programming language",
            "category": "programming"
        }
    ),
    PointStruct(
        id=6,
        vector=[0.0, 0.8, 0.7, 0.6],
        payload={
            "text": "Cricket is popular in India",
            "category": "sports"
        }
    ),
    PointStruct(
        id=3,
        vector=[0.12, 0.22, 0.31, 0.41],
        payload={
            "text": "Django is a Python framework",
            "category": "programming"
        }
    )
]

client.upsert(
    collection_name="documents",
    points=points
)

print("Data inserted!")
```

```python
from qdrant_client.models import Filter, FieldCondition, MatchValue

results = client.query_points(
    collection_name="documents",
    query=[0.11, 0.21, 0.32, 0.42],
    query_filter=Filter(
        must=[
            FieldCondition(
                key="category",
                match=MatchValue(value="programming")
            )
        ]
    ),
    limit=5
)
```

```python
results = client.query_points(
    collection_name="documents",
    query=[0.11, 0.21, 0.32, 0.42],
    limit=3
)

for point in results.points:
    print(point.id)
    print(point.score)
    print(point.payload)
    print()
```
